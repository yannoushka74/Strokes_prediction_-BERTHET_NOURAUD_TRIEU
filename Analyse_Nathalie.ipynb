```python
import pandas as pd
import numpy as np

df = pd.read_csv("D:\Documents\GitHub\Strokes_prediction_BERTHET_NOURAUD_TRIEU\strokes.csv", sep=',', index_col='id')
# Exploration des données
print("- Nombre d'observations et de colonnes\n: ", df.shape,'\n')
print("- Colonnes\n: ", df.info(),'\n')
print("- Stats 'age' :\n", df['age'].describe(),'\n')
```

```python
print("Gender :\n", df['gender'].value_counts(normalize=True),'\n')
print("hypertension :\n", df['hypertension'].value_counts(normalize=True),'\n')
print("work_type :\n", df['work_type'].value_counts(normalize=True),'\n')
print("stroke :\n", df['stroke'].value_counts(normalize=True),'\n')
```

```python
# Doublons
print('- Nombre de doublons :', df.duplicated().sum())


# Détection des valeurs manquantes
print('- Variables ayant des valeurs manquantes :')
df.isna().any(axis = 0)

```

```python
# Classe d'âge
df['age_classes'] = pd.qcut(df['age'], labels=[0, 1, 2, 3], q=4)
# répartition des attaques selon la classe d'âge
df_ca = df.groupby(['age_classes', 'stroke']).agg({'stroke' : 'count'})
df_ca.head(10)
```

```python
# répartition des attaques selon type de travail
df_ca = df.groupby(['work_type', 'stroke']).agg({'work_type' : 'count'})
df_ca.head(20)
```

```python
# répartition des attaques selon le genre
df_ca = df.groupby(['gender', 'stroke']).agg({'stroke' : 'count'})
df_ca.head(10)
```

```python
# répartition des attaques selon l'hypertension
df_ca = df.groupby(['hypertension', 'stroke']).agg({'stroke' : 'count'})
df_ca.head(10)
```
